```python
from pathlib import Path
import pandas as pd
```

```python
item_list_2016_csv = Path("./mndot_bid_etl/data/csv/item_list_2016.csv").resolve()
item_list_2018_csv = Path("./mndot_bid_etl/data/csv/item_list_2018.csv").resolve()
item_list_2020_csv = Path("./mndot_bid_etl/data/csv/item_list_2020.csv").resolve()
```

```python
read_csv_2016_kwargs = {
    "dtype": "str",
    "quotechar": "'",
    "header": "infer",
    # "skiprows": 3713,
    # "nrows": 11,
}

item_list_2016 = pd.read_csv(item_list_2016_csv, **read_csv_2016_kwargs)
item_list_2016.head(11)
```

```python
pipe_aprons_2016 = item_list_2016.iloc[3712:3723]
pipe_aprons_2016.head(20)
```

```python
from mndot_bid_etl.transform.item_list import transform_item_df
```

```python
transformed_pipe_aprons_2016 = transform_item_df(pipe_aprons_2016)
transformed_pipe_aprons_2016.head(20)
```

```python
transformed_pipe_aprons_2016.at[3712, "long_description"]
```

```python
def generate_compound_ids(df: pd.DataFrame) -> pd.Series:
    return df["id"] + "_" + df["long_description"]
```

```python
compound_ids_2016 = generate_compound_ids(transformed_pipe_aprons_2016)
compound_ids_2016
```

```python
compound_ids_2016[3712]
```

```python
df_2016 = pd.DataFrame(index=compound_ids_2016).assign(is_in_2016=True)
df_2016.head()
```

```python
with open("./df_2016_dump.csv", "w") as f:
    df_2016.to_csv(f, quotechar="'")
```
